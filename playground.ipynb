```python
%load_ext autoreload
%autoreload 2
from bandit import *
```

```python
bandit = Bandit(k=10, mu=0, sigma=1)
agents = [Agent(10) for _ in range(10)]
agents
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
for i, agent in enumerate(agents):
    epsilon = 1/(i+1)
    agent.play(bandit, max_iter=1000, epsilon=epsilon)
    ax.plot(np.arange(1001), agent.average_wealth, label=str(epsilon))
ax.legend()
```

```python
plt.scatter(x=np.arange(10), y=bandit.expected_rewards, label='expected reward')
plt.scatter(x=np.arange(10), y=agents[6].estimates, color='red', label='learned estimate')
```
